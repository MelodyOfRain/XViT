```python
import torch
from sklearn.metrics import f1_score,hamming_loss

def metric_func(data,output):
    pred = torch.sigmoid(output).detach().cpu().numpy()
    score = f1_score(data[-1].cpu().numpy(),(pred>0.5).astype(int),average='macro',zero_division=0)
    # score = f1_score(data[-1].cpu().numpy(),(pred>0.5).astype(int),average='micro',zero_division=0)
    # score = hamming_loss(data[-1].cpu().numpy(),(pred>0.5).astype(int))
    return score
```
